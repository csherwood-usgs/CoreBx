```python
# Example of converting date strings to datetime

import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
%matplotlib inline

dates = ([\
         "30 Aug 2019",\
         "13 Sep 2019",\
         "11 Oct 2019",\
         "26 Nov 2019"])

dtl = []
for date in dates:
    dtl.append(datetime.strptime(date, "%d %b %Y"))
dt = np.array(dtl)

# copied from stats.txt:
Vsum = np.array([14891535.25233719, 12223653.37997114, 11400367.25328069, 11724666.57830751])
Vsumall = np.array([18098349.80286515, 16026037.01528873, 13951215.04976664, 14615301.54792139])

fig, ax = plt.subplots(1)
fig.autofmt_xdate()
plt.plot(dt,Vsum/1e6)
plt.plot(dt,Vsumall/1e6)
plt.ylabel('Million m$^3$')
```
