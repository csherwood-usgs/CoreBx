```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

trans = np.array((1,2,2,2,1,3,4,1))
x =     np.array((1,2,3,4,5,6,7,8))
cz = np.mean(trans)
cx = np.sum(trans*x)/np.sum(trans)

print("cz, cx:",cz,cx)
```

```python
plt.plot(x,trans)
plt.plot(cx,cz,'o')
```

```python
trans = np.array((1,np.nan,2,2,np.nan,3,4,1))
x =     np.array((1,2,3,4,5,6,7,8))
cz = np.nanmean(trans)
cx = np.nansum(trans*x)/np.nansum(trans)

print("cz, cx:",cz,cx)
plt.plot(x,trans)
plt.plot(cx,cz,'o')
```
