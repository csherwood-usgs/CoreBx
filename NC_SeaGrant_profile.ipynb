```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd

from scipy import interpolate, signal, stats

%matplotlib inline
```

```python
df=pd.read_csv("D:\\crs\\docs\\papers\\2020_NC_SeaGrant\\profile.csv",header = 0)
#print(df)
print("Read: ",len(df.index),' lines.')
dist = np.array(df['Distance'])
Aug = np.array(df['Aug'])
Sep = np.array(df['Sep'])
Oct = np.array(df['Oct'])
Nov = np.array(df['Nov'])
```

```python
#cols=['#fef0d9','#fdcc8a','#fc8d59','#d7301f']
cols=['#feedde','#fdbe85','#fd8d3c','#e6550d','#a63603']
```

```python
plt.plot(dist,np.ones_like(dist)*0.4,'--',c='dimgray',linewidth=2)
plt.plot(dist,Aug,linewidth=3,c=cols[1],label='28-Aug')
plt.plot(dist,Sep,linewidth=3,c=cols[2],label='13-Sep')
plt.plot(dist,Oct,linewidth=3,c=cols[3],label='11-Oct')
plt.plot(dist,Nov,linewidth=3,c=cols[4],label='26-Nov')
plt.legend()
plt.ylabel('Elevation (m NAVD88)')
plt.xlabel('Across-shore Distance (m)')
plt.savefig('profile.svg')
```
